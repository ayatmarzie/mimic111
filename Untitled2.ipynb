```python
from transformers import AutoTokenizer
```

```python
import pandas as pd
```

```python
tokenizer=AutoTokenizer.from_pretrained('medicalai/ClinicalBERT')
```

```python
input_file='final_text.csv'
output_file='embeddings.csv'
batch_size=1
first_batch=True
for chunk in pd.read_csv(input_file,chunksize=batch_size):
    batch_texts=chunk['text'].tolist()
    tokenized_batch=tokenizer(batch_texts)
    chunk['input_ids']=[str(ids) for ids in tokenized_batch['input_ids']]
    chunk['attention_mask']=[str(masks) for masks in tokenized_batch['attention_mask']]
    chunk.to_csv(output_file,mode='a',index=False,header=first_batch)
    first_batch=False
```
